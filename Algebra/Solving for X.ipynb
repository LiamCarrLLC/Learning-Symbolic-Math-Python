```python
from IPython.display import display, Math
import sympy as sym

sym.init_printing()
```

```python
x, y = sym.symbols('x, y')

# 9 = 2x + 4
expr = 2 * x + 4 - 9
```

```python
display(Math('\\text{The solution to } (%s) \\text{ is:     x =} %s' % (sym.latex(expr), sym.latex(sym.solve(expr)))))
```

```python
expr = x ** 2 - 4
sol = sym.solve(expr)

for i in range(0, len(sol)):
    print('Solution ' + str(i) + ' is ' + str(sol[i]))
```

```python
expr = x / 4 - x * y + 5
display(Math(sym.latex(expr)))
```

```python
display(Math(sym.latex(sym.solve(expr, x))))
```

```python
display(Math(sym.latex(sym.solve(expr, y))))
```

```python
m, x, y, b = sym.symbols('m,x,y,b')
slope_intercept_formula = m * x - y + b

display(Math(sym.latex(slope_intercept_formula)))
```
