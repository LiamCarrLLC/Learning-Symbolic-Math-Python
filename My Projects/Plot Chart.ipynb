```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math

sym.init_printing()
```

```python
def draw_grid():
    axis = plt.gca()
    plt.plot(axis.get_xlim(), [0, 0], 'k--')
    plt.plot([0, 0], axis.get_ylim(), 'k--')


def draw_point(x, y):
    plt.plot(x, y, 'o')


def draw_points(point_list):
    for point in point_list:
        plt.plot(point[0], point[1], 'o')


def draw_function(x_range, f):
    x = np.linspace(x_range[0], x_range[1], 100)
    y = f(x)
    plt.plot(x, y)


slope_intercept_form = lambda x: 3 / 2 * x + 1

draw_point(1, 1)
draw_points([[2, 2]])
draw_function([-5, 5], slope_intercept_form)

draw_grid()
plt.show()
```
