```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
from IPython.display import display, Math

sym.init_printing()
```

```python
y = np.array([[1, 1], [2, 3], [3, 1]])

p = Polygon(y, facecolor='g', alpha=0.3)

fig, ax = plt.subplots()
ax.add_patch(p)

ax.set_xlim([0, 4])
ax.set_ylim([0, 4])
plt.show()
```

```python
x = np.linspace(-2, 2, 101)
f = -x ** 2
f2 = x ** 3

pnts = np.vstack((x, f)).T
pnts2 = np.vstack((x, f2)).T

p = Polygon(pnts, facecolor='g', alpha=.2)
p2 = Polygon(pnts2, facecolor='r')

fig, ax = plt.subplots()

plt.plot(x, f, 'k')
plt.plot(x, f2, 'k')

ax.fill_between(x, f, f2)

plt.savefig('firstfig.pdf')

plt.show()
```

```python
x = np.linspace(-2, 2, 101)
x2 = np.linspace(-5, 5, 101)

f = 3 / 2 * x + 5
f2 = x * 0

fig, ax = plt.subplots()

plt.plot(x, f, 'b')
plt.plot(x2, f2, 'r')

ax.fill_between(x, f, f2, facecolor='y', alpha=.1)

ax.set_xlim([-10, 10])
ax.set_ylim([-5, 10])

plt.show()
```

```python
f = 3 * x + 5
f2 = x * 0 - 20

fig, ax = plt.subplots()

ax.fill_between(x, f, f2, facecolor='y', alpha=.1)

plt.plot(x, f)

axis = plt.gca()
plt.plot(axis.get_xlim(), [0, 0], 'k--')
plt.plot([0, 0], axis.get_ylim(), 'k--')

plt.show()
```
