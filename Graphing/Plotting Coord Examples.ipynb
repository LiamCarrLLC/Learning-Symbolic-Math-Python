```python
import sympy as sym
from IPython.display import display, Math
import matplotlib.pyplot as plt
sym.init_printing()
```

```python
x, y = sym.symbols('x,y')
y = x**2 - 3 * x

display(Math(sym.latex(y)))
```

```python
for i in range(-10, 12):
    y_coord = y.subs(x, i)
    plt.plot(i, y_coord, 'o')

plt.xlabel('x')
plt.ylabel('f(x) = x^2 - 3x')

plt.grid()
plt.show()
```
