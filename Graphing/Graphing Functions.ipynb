```python
import sympy as sym
import sympy.plotting.plot as symplot
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math

sym.init_printing()
```

```python
x, y = sym.symbols('x, y')

# x1 = range(-3, 4)
x = np.linspace(-3, 4, 30)
y = np.zeros(len(x))

for i in range(0, len(x)):
    y[i] = 2 - x[i] ** 2

plt.plot(x, y, '-')
```

```python
x = np.linspace(-3, 4, 100)
y = 2 + np.sqrt(abs(x))

plt.plot(x, y, 'm-')
```

```python
x, y = sym.symbols('x, y')

y = x ** 2

# Global plot attributes
p = symplot(y, show=False)
p.xlim = [0, 50]
p.title = "$y=x^2$"

# Line one attributes
p[0].line_color = 'm'

p.show()
```

```python
x, a = sym.symbols('x,a')

expr = a / x
n = 1

p = symplot(expr.subs(a, n), (x, -5, 5), show=False)
p.ylim = [-5, 5]
p[0].label = '$y=%s$' % sym.latex(expr.subs(a, n))

p2 = symplot(expr.subs(a, 2), show=False)
p.extend(p2)
p[1].label = '$y=%s$' % sym.latex(expr.subs(a, 2))

p.legend = True
p.show()
```
