```python
# y = mx + b

# x1 and x2 values
x = [-5, 5]
m = 2
b = 1

# evaluate one point at a time from a list of x coordinates
# y = [340, 2345]
# for i in range(0, len(x)):
#     y[i] = m*x[i] + b

# Convert x into a numpy array to use the formula as is on the list of x coordinates to generate a list on y coordinates
y = m * np.array(x) + b

plt.plot(x, y, label='y=%sx+%s' % (m, b))

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

# Draw the x and y axis
axis = plt.gca()
plt.plot(axis.get_xlim(), [0, 0], 'k--')
plt.plot([0, 0], axis.get_ylim(), 'k--')

plt.legend()
plt.show()
```

```python
m = [.7, -1.25]
x = [-5, 5]
b = [-2, .75]

for i in range(0, len(m)):
    y = m[i] * np.array(x) + b[i]
    plt.plot(x, y, label='y=%sx+%s' % (m[i], b[i]))

# Use sympy to plot the point of intersection (algebra equation)
plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

# Draw the x and y axis
axis = plt.gca()
plt.plot(axis.get_xlim(), [0, 0], 'k--')
plt.plot([0, 0], axis.get_ylim(), 'k--')

plt.legend()
plt.show()
```

```python
b, m, x, y = sym.symbols('b, m, x,y')

m = -1
b = 900

slope_intercept_formula = m * x - y + b

x_intercept_formula = sym.solve(slope_intercept_formula, x)[0]
y_intercept_formula = sym.solve(slope_intercept_formula, y)[0]

x_intercept = sym.solve(y_intercept_formula.subs({m: m, x: x, b: b}))
y_intercept = sym.solve(x_intercept_formula.subs({m: m, y: y, b: b}))

display(Math('x=%s' % sym.latex(x_intercept_formula)))
display(Math('y=%s' % sym.latex(y_intercept_formula)))

plt.plot([0, y_intercept[0]], [x_intercept[0], 0], label='y =%sx + %s' % (m, b))

# Use sympy to plot the point of intersection (algebra equation)
plt.axis('square')
plt.grid()
plt.xlim(int(-x_intercept[0]), int(x_intercept[0]))
plt.ylim(int(-x_intercept[0]), int(x_intercept[0]))

# Draw the x and y axis
axis = plt.gca()
plt.plot(axis.get_xlim(), [0, 0], 'k--')
plt.plot([0, 0], axis.get_ylim(), 'k--')

plt.legend()
plt.show()
```
