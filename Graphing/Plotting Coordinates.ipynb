```python
import sympy as sym
import matplotlib.pyplot as plt
sym.init_printing()
```

```python
x = 3
y = 5
plt.plot(x, y, 'ro')
plt.axis('square')
plt.axis([-6, 6, -6, 6])
plt.grid()

plt.show()
```

```python
x = [-4, 5, 2, 4, 5, 6, 5]
y = [0, 3, 5, 2, -3, 4, 8]

for i in range (0, len(x)):
    plt.plot(x[i], y[i], 'o', label='point %i' %(i+1))

plt.axis('square')
plt.axis([-4.5, 8.5, -3.5, 8.5])
plt.grid()
plt.legend()
plt.show()
```
